```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
county_cancer = pd.read_csv("data/processed/county_cancer.csv", index_col=[0])
county_cancer.head(3)
```

```python
county_cancer.shape
```

```python
county_chemicals = pd.read_csv("data/processed/county_chemicals.csv")
county_chemicals.head(3)
```

```python
county_rivers = pd.read_csv("data/processed/county_rivers.csv")
county_rivers.head(3)
```

```python
full_county_dataset = county_cancer.merge(county_chemicals, on = "fips", how = "left")
full_county_dataset.head(3)
```

```python
print(county_cancer.shape)
print(county_chemicals.shape)
print(full_county_dataset.shape)
```

```python
full_county_dataset.isnull().sum()
```

# Merge the rivers dataset to full county data

```python
full_county_dataset = full_county_dataset.merge(county_rivers, 
                                                left_on="fips", 
                                                right_on="profile_STATECTY",
                                                how="left")
```

```python
full_county_dataset.shape
```

```python
print(full_county_dataset.isnull().sum())
```

```python
plt.figure(figsize = (12,12))
sns.heatmap(full_county_dataset.isnull());
# Missing values
```

```python
full_county_dataset.to_csv("data/processed/full_county_dataset.csv", index=False)
```
