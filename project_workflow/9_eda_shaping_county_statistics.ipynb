# County Statistics

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
counties_2019 = pd.read_csv("data/raw/county_2019.csv")
county_amer_com_surv = pd.read_csv("data/raw/county_info_openintro.csv")
counties_2019.head()
```

```python
counties_2019.columns
```

```python
county_amer_com_surv.head()
```

```python
county_fips = counties_2019[["name", "fips", "age_over_65"]]
county_fips.head()
```

```python
print(county_fips.shape)
print(county_amer_com_surv.shape)
```

```python
np.mean(county_fips["name"] == county_amer_com_surv["name"]) #looks like these counties aren't quite exactly matched.
```

```python
county_fips[county_fips["name"] != county_amer_com_surv["name"]]
```

```python
county_amer_com_surv[county_amer_com_surv["name"] != county_fips["name"]]
```

```python
# since these two county lists are identical except for punctiation, we will concat them rather than left join.
county_data = pd.concat([county_fips, county_amer_com_surv.drop(columns = "name")], axis = 1)
county_data
```

```python
county_data.to_csv("data/processed/county_information.csv", index = False)
```
