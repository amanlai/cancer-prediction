```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns" , 100)
```

```python
riv_2013 = pd.read_csv("data/cleaned/2013_river_site_hab_chem_bminv.csv")
riv_2018 = pd.read_csv("data/cleaned/2018_river_site_hab_chem_bminv.csv")
```

```python
cols_2013 = set(riv_2013.columns)
cols_2018 = set(riv_2018.columns)
cols_2013.intersection(cols_2018)
```

```python
riv_2013 = riv_2013[[column for column in riv_2013.columns if column in riv_2018.columns]]
riv_2013
```

```python
riv_2018 = riv_2018[[column for column in riv_2018.columns if column in riv_2013.columns]]
riv_2018
```

```python
riv_concat = pd.concat([riv_2013, riv_2018], )
riv_concat.head()
```

```python
print(riv_2013.shape)
print(riv_2018.shape)
print(riv_concat.shape)
```

```python
riv_concat["profile_STATECTY"].value_counts().hist()
```

```python
riv_concat_obj = ["profile_STATECTY", "profile_CNTYNAME"]
riv_concat_num = list(riv_concat.select_dtypes("number").columns)
columns = riv_concat_obj + riv_concat_num
#columns
```

```python
def agg_func(df):
    aggs = {}
    for col in df.columns:
        if col in riv_concat_obj:
             aggs[col] = [pd.Series.mode]
        if col in riv_concat_num:
            aggs[col] = [pd.Series.mean]
    return aggs

riv_concat = riv_concat.groupby(["profile_STATECTY"]).agg(agg_func(riv_concat)).round(2).reset_index()
riv_concat.columns = riv_concat.columns.droplevel(1) 
```

```python
riv_concat = riv_concat.loc[:,~riv_concat.columns.duplicated()].drop(columns = "profile_STATECTY.1")
riv_concat.head()
```

```python
for col in riv_concat.select_dtypes("number").columns:
    riv_concat[col] = riv_concat[col].fillna(riv_concat[col].mean())
```
